## 课程框架

```python
import pandas as pd

balance_df = pd.read_excel('资产负债表.xlsx')
balance_df.head()
```

```python
income_df = pd.read_excel('利润表.xlsx')
income_df.head()
```

```python
cashflow_df = pd.read_excel('现金流量表.xlsx')
cashflow_df.head()
```

```python
data = pd.merge(balance_df, income_df, on=['TS 股票代码', '名称', '行业名称', '报告日期'])
data.head()
```

```python
data = pd.merge(data, cashflow_df, on=['TS 股票代码', '名称', '行业名称', '报告日期'])
data.info()
```
